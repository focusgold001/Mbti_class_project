# 학생들의 MBTI답변 유사성 검사

```python
import pymysql
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity


# 원하는 유저의 유사도 조회하는 함수

def cosine(num, num2):
    conn = pymysql.connect(user='root', password='qwer1234', host='127.0.0.1', db='project', charset='utf8')
    sql_input = "SELECT * FROM testdata"
    df_this_data = pd.read_sql_query(sql_input, conn)
    dj=pd.DataFrame(cosine_similarity(df_this_data, dense_output=True))
    result = dj.loc[num][num2]
    return result


student_num = [i for i in range(500)] # 1부터 500까지의 숫자를 변수에 저장

from itertools import permutations # 모듈 불러오기


i = 0 # 숫자를 넣을 변수



for mate in permutations(student_num, 25): # 학생 500명을 25명씩 나눠보는 경우의수(조합)
    total = 0  # 유사도 총점을 넣을 변수
    i = i+1 # 반복문이 한번 완료될때마다 1씩 증가
    if i==21: # i가 21이되면 멈추기
        break
    #rint(i,'반',mate)

    for two in permutations(mate, 2): # 둘씩 묶기
        result = cosine(two[0], two[1]) # 코사인 함수 이용해서 순서대로 두명의 유사도 비교하기
        total = total + result # 유사도 총점을 total에 저장
       
    print(i, "번째 계산된 반", mate, "유사도 총점 : ", total)
    print()
```
